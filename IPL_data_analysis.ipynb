```python
#Slicing through virat's dissmissal data
virat_dismissal=ipl_deliveries[ipl_deliveries['player_dismissed']=='V Kohli']
virat_dismissal_original=virat_dismissal
virat_dismissal_original
virat_max=virat_dismissal_original.head(10)
virat_dismissal_original.head(10)
```

```python
#A Nehra took viratkohli's wicket for almost 6 times ...from groupby method
number_of_dismissal=virat_dismissal_original.groupby('bowler')['player_dismissed'].count().sort_values(ascending=True)
number_of_dismissals=pd.DataFrame(number_of_dismissal)
number_of_dismissals
```
